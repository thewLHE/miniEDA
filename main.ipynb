```python
import plotly.graph_objects as go

def show_graph(df):
    fig = go.Figure()
    
    # 각 열에 대해 한 개의 바(trace)를 추가
    for col in df.columns:
        data = df[col].to_list()  # Polars의 열 데이터를 리스트로 변환
        fig.add_trace(go.Bar(x=[col] * len(data), y=data, name=col))
    
    # 그래프 레이아웃 업데이트
    fig.update_layout(
        title="Bar Graph of Each Column",
        xaxis_title="Columns",
        yaxis_title="Values",
        barmode='group',  # 그룹 모드로 바를 그룹화
        xaxis_tickangle=-45  # X축 라벨 각도 조정
    )
    
    # 그래프 표시
    fig.show()
```

```python
# 그래프 표시
show_graph(df_csv)
```

```python
from highcharts_core.chart import Chart
import polars as pl

# 데이터 준비
data = pl.DataFrame({
    'angle': [0, 45, 90, 135, 180, 225, 270, 315, 360],
    'value': [1, 2, 3, 4, 5, 4, 3, 2, 1]
})

# 차트 객체 생성
chart = Chart()

# 차트의 기본 구성 설정
chart.options = {
    'chart': {'polar': True, 'type': 'line'},
    'title': {'text': 'Polar Chart Example'},
    'pane': {'startAngle': 0, 'endAngle': 360},
    'xAxis': {
        'categories': data['angle'].cast(str).to_list(),  # Convert categories to strings
        'tickInterval': 45,
        'min': 0,
        'max': 360,
        'labels': {'format': '{value}°'}
    },
    'yAxis': {'min': 0} }
```

## show

```python
from highcharts_core.chart import Chart
import polars as pl
import json


# 데이터 추가
chart.add_series({
    'type': 'line',  # Add the type key here
    'name': 'Value',
    'data': data['value'].to_list(),
    'pointPlacement': 'on'
})

# 차트 옵션을 딕셔너리로 변환
chart_options_dict = chart.options.to_dict()

# 차트 옵션을 JSON으로 변환
chart_options_json = json.dumps(chart_options_dict)

# HTML 코드 생성
html_code = f"""
<!DOCTYPE html>
<html>
<head>
    <script src="https://code.highcharts.com/highcharts.js"></script>
    <script src="https://code.highcharts.com/highcharts-more.js"></script>
    <script src="https://code.highcharts.com/modules/exporting.js"></script>
    <script src="https://code.highcharts.com/modules/accessibility.js"></script>
</head>
<body>
    <div id="container" style="width:100%; height:400px;"></div>
    <script>
        document.addEventListener('DOMContentLoaded', function () {{
            var chartOptions = {chart_options_json};
            Highcharts.chart('container', chartOptions);
        }});
    </script>
</body>
</html>
"""

# HTML 파일로 저장
with open('polar_chart.html', 'w') as file:
    file.write(html_code)
```
